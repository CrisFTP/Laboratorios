````python
def nota_chilena(dataframe_in, columna):
    # Copiar el DataFrame de entrada para evitar modificar el original
    dataframe_out = dataframe_in.copy()

    # Aplicar la ecuaci√≥n de la escala chilena a la columna especificada
    dataframe_out[columna] = (dataframe_in[columna] * 0.06) + 1

    # Redondear las notas transformadas al segundo decimal
    dataframe_out[columna] = dataframe_out[columna].round(2)

    return dataframe_out
````

````python
def alumnos_promedio(dataframe_in):
    # Copiar el DataFrame de entrada para evitar modificar el original
    dataframe_out = dataframe_in.copy()

    # Calcular el promedio de cada alumno
    promedio_alumno = dataframe_out[['math score', 'reading score', 'writing score']].mean(axis=1)

    # Redondear el promedio al segundo decimal
    promedio_alumno = promedio_alumno.round(2)

    # Agregar el promedio como una nueva columna al DataFrame original
    dataframe_out['GPA'] = promedio_alumno

    return dataframe_out

df_grades=alumnos_promedio(df_grades)
display(df_grades)
````

````python
def reprobados_destacados(dataframe_in):
    # Obtener DataFrame con los alumnos que reprobaron el a√±o
    alumnos_reprobados = dataframe_in[dataframe_in['GPA'] < 4.0].reset_index(drop=True)

    # Obtener DataFrame similar al original, pero marcando con una "R" el atributo GPA de los alumnos reprobados
    dataframe_out = dataframe_in.copy()
    dataframe_out.loc[dataframe_out['GPA'] < 4.0, 'GPA'] = 'R'

    # Obtener DataFrame con los alumnos destacados ordenados de mayor a menor
    alumnos_destacados = dataframe_in[(dataframe_in['math score'] > 5.8) &
                                       (dataframe_in['reading score'] > 5.8) &
                                       (dataframe_in['writing score'] > 5.8)]
    alumnos_destacados = alumnos_destacados.sort_values(by='GPA', ascending=False).reset_index(drop=True)

    return alumnos_reprobados, dataframe_out, alumnos_destacados
````

````python
# C√≥digo para quintiles

def generar_quintiles(dataframe_in):
    # Generar quintiles de rendimiento para los alumnos seg√∫n su promedio general
    quintiles = pd.qcut(dataframe_in['GPA'], q=5, labels=['Baj√≠simo', 'Bajo', 'Medio', 'Alto', 'Muy alto'])

    # Agregar la columna de quintiles al DataFrame
    dataframe_in['GPA cuantiles'] = quintiles

    return dataframe_in

df_grades= generar_quintiles(df_grades)
"""
NOTA: La variable del dataframe que contenga los quintiles debe llamarse
'GPA cuantiles' para que el gr√°fico funcione.
"""
````

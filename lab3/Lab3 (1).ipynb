### 3.1 Transformación de notas a "Nota Chilena" [0.3 Puntos]

En una de las reuniones con Don Caguayo este les comenta su intención de pasar todas sus notas a la "Escala Chilena". Para esto les facilita la siguiente ecuación:

$$Nota \, Chilena= \% \, de \, logro*0.06 + 1$$

Utilizando la ecuación entregada, cree una función que le permita pasar a la escala Chilena las notas del dataframe y entregue las notas redondeadas al segundo decimal.

**To-Do**

- [ ] Tranforme los valores a la escala chilena.

**Respuesta:**

````python
def nota_chilena(dataframe_in, columna):
    # Copiar el DataFrame de entrada para evitar modificar el original
    dataframe_out = dataframe_in.copy()

    # Aplicar la ecuación de la escala chilena a la columna especificada
    dataframe_out[columna] = (dataframe_in[columna] * 0.06) + 1

    # Redondear las notas transformadas al segundo decimal
    dataframe_out[columna] = dataframe_out[columna].round(2)

    return dataframe_out
````

### 3.2 Obtención de promedio de alumnos [0.3 Puntos]

Con todas las notas adjuntadas, la prestigiosa Universidad De la Cachaña necesita obtener el promedio general de cada uno de los alumnos. **To-Do:**
- [ ] Crear función que nos permite obtener el promedio de cada uno de los alumnos y anexar esta información en el `DataFrame` original.


**Respuesta:**

````python
def alumnos_promedio(dataframe_in):
    # Copiar el DataFrame de entrada para evitar modificar el original
    dataframe_out = dataframe_in.copy()

    # Calcular el promedio de cada alumno
    promedio_alumno = dataframe_out[['math score', 'reading score', 'writing score']].mean(axis=1)

    # Redondear el promedio al segundo decimal
    promedio_alumno = promedio_alumno.round(2)

    # Agregar el promedio como una nueva columna al DataFrame original
    dataframe_out['GPA'] = promedio_alumno

    return dataframe_out

df_grades=alumnos_promedio(df_grades)
display(df_grades)
````

````python
def reprobados_destacados(dataframe_in):
    # Obtener DataFrame con los alumnos que reprobaron el año
    alumnos_reprobados = dataframe_in[dataframe_in['GPA'] < 4.0].reset_index(drop=True)

    # Obtener DataFrame similar al original, pero marcando con una "R" el atributo GPA de los alumnos reprobados
    dataframe_out = dataframe_in.copy()
    dataframe_out.loc[dataframe_out['GPA'] < 4.0, 'GPA'] = 'R'

    # Obtener DataFrame con los alumnos destacados ordenados de mayor a menor
    alumnos_destacados = dataframe_in[(dataframe_in['math score'] > 5.8) &
                                       (dataframe_in['reading score'] > 5.8) &
                                       (dataframe_in['writing score'] > 5.8)]
    alumnos_destacados = alumnos_destacados.sort_values(by='GPA', ascending=False).reset_index(drop=True)

    return alumnos_reprobados, dataframe_out, alumnos_destacados
````

### 3.4 Quintiles de Rendimiento [0.5 Puntos]

El rector, intrigado por saber si existe una relación entre el nivel educacional de los padres y las notas de sus hijos les plantea estudiar dicha relación. Para esto, les solicita que desde el `DataFrame` que posee a todos los alumnos (`DataFrame` obtenido tras la limpieza), generen una nueva columna que clasifique en quintiles de rendimiento para sus alumnos: `['Bajísimo', 'Bajo', 'Medio', 'Alto', 'Muy alto']`. (investigar `qcut()`)

Hecho esto, grafique a través de un scatter plot el nivel educacional de los padres v/s el promedio de los alumnos, utilice los cuantiles generados anteriormente para mostrar en que cuantil se encuentra cada uno de los datos.

Realizado el grafico, responda la siguiente pregunta: **¿Es posible visualizar alguna tendencia en el gráfico?.**



**Respuesta:**

````python
# Código para quintiles
def generar_quintiles(dataframe_in):
    # Generar quintiles de rendimiento para los alumnos según su promedio general
    quintiles = pd.qcut(dataframe_in['GPA'], q=5, labels=['Bajísimo', 'Bajo', 'Medio', 'Alto', 'Muy alto'])

    # Agregar la columna de quintiles al DataFrame
    dataframe_in['GPA cuantiles'] = quintiles

    return dataframe_in

df_grades= generar_quintiles(df_grades)
````

````python
# Inserte su respuesta aquí

#Aplicamos la funcion de la 3.2 para obtener columna con promedios
data_unida = alumnos_promedio(data_unida)

# Definimos un diccionario para el posterior mapeo
mapeo_educacion = {
    'some high school': 'school',
    'some college': 'school',
    'high school': 'school',
    "bachelor's degree": 'college',
    "associate's degree": 'college',
    "master's degree": 'postgraduate'
}

# Aplicamos el diccionario con .map
data_unida['simple parental level of education'] = data_unida['parental level of education'].map(mapeo_educacion)

#Agregamos el promeido de notas según race/ethnicity y simple parental level of education
promedio_notas = data_unida.groupby(['race/ethnicity', 'simple parental level of education']).agg({'math score': 'mean', 'reading score': 'mean', 'writing score': 'mean', 'GPA': 'mean'}).reset_index()

# Agregamos el conteo de alumnos según race/ethnicity y simple parental level of education
conteo_alumnos = data_unida.groupby(['race/ethnicity', 'simple parental level of education']).size().reset_index(name='count')

# Calculamos el porcentaje respecto al total, lo hacemos string y le agregamos el '%'
total_alumnos = conteo_alumnos['count'].sum()
porcentaje_alumnos = (conteo_alumnos['count'] / total_alumnos) * 100
conteo_alumnos['percentage'] = porcentaje_alumnos.round(2).astype(str) + ' %'

# Fusionamos con merge el promedio de notas hecho con el de conteo
tabla_final = promedio_notas.merge(conteo_alumnos, on=['race/ethnicity', 'simple parental level of education'])
# Ordenamos con .sort
tabla_final = tabla_final.sort_values(by=['race/ethnicity', 'simple parental level of education']).reset_index(drop=True)

#Mostramos la tabla final
display(tabla_final)
````
